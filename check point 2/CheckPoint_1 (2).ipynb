### **Outliers Values**

Dibagian ini kita akan melihat nilai data yang ekstrem, hal ini dilakukan untuk mendapatkan insight yang akurat sehingga dapat memperkuat analisis.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Hasil Indikasi Outlier:
Beberapa variabel menunjukkan perbedaan besar antara mean dan median, terutama:

total sulfur dioxide

residual sugar

chlorides

sulphates

Selain itu, nilai maksimum beberapa variabel cukup tinggi dibanding median.

Insight & Penjelasan:
Outlier kemungkinan terjadi karena:

Perbedaan metode produksi wine

Jenis wine berbeda

Variasi fermentasi

Kondisi kimia ekstrem

Contoh:

total sulfur dioxide max = 289 (cukup jauh dari median 38)

residual sugar max = 15.5 (jauh di atas median 2.2)

Outlier dapat:

Mempengaruhi nilai mean

Meningkatkan standar deviasi

Mempengaruhi model regresi

Namun, dalam data kimia, outlier belum tentu kesalahan — bisa jadi karakteristik unik suatu wine.

Kesimpulan:
Dataset memiliki beberapa nilai ekstrem (outlier), tetapi masih dalam batas logis dan tidak menunjukkan kesalahan input.

Aktivitas

Melihat komposisi rata-rata kandungan kimia dalam dataset untuk memahami karakteristik umum wine.
