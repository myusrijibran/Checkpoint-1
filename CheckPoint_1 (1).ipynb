### **Outliers Values**

Dibagian ini kita akan melihat nilai data yang ekstrem, hal ini dilakukan untuk mendapatkan insight yang akurat sehingga dapat memperkuat analisis.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Hasil Indikasi Outlier:
Beberapa variabel menunjukkan perbedaan besar antara mean dan median, terutama:

total sulfur dioxide

residual sugar

chlorides

sulphates

Selain itu, nilai maksimum beberapa variabel cukup tinggi dibanding median.

Insight & Penjelasan:
Outlier kemungkinan terjadi karena:

Perbedaan metode produksi wine

Variasi fermentasi

Kondisi kimia ekstrem

Contoh:

Outlier dapat:

Namun, dalam data kimia, outlier belum tentu kesalahan — bisa jadi karakteristik unik suatu wine.

Kesimpulan:
Dataset memiliki beberapa nilai ekstrem (outlier), tetapi masih dalam batas logis dan tidak menunjukkan kesalahan input.

## **Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Box Plot**

```python
VARIABEL1 = df['quality'].value_counts().head(5).index
VARIABEL2 = df[df['quality'].isin(VARIABEL1)]

plt.figure(figsize=(12, 7)) # Ukuran figure
sns.boxplot(x='quality', y='alcohol', data=VARIABEL2, palette='viridis', order=VARIABEL1, hue='quality', legend=False)
plt.title('Distribusi Alkohol berdasarkan Tingkat Kualitas Wine', fontsize=16) # Judul
plt.xlabel('Tingkat Kualitas Wine', fontsize=12) # X Keterangan
plt.ylabel('Kadar Alkohol', fontsize=12) # Y Keterangan
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Wine dengan kualitas lebih tinggi cenderung memiliki kadar alkohol yang lebih tinggi.
Terlihat bahwa median kadar alkohol meningkat seiring dengan naiknya tingkat kualitas wine. Misalnya, kualitas 7 dan 8 memiliki median alkohol yang lebih tinggi dibandingkan kualitas 4 dan 5. Terdapat beberapa outlier pada kategori kualitas tertentu. Beberapa titik di luar box menunjukkan adanya wine dengan kadar alkohol yang jauh lebih tinggi atau lebih rendah dibandingkan mayoritas data dalam kategori tersebut. Tindakan yang dapat dilakukan adalah mengoptimalkan proses fermentasi untuk meningkatkan kualitas wine.
Karena kadar alkohol cenderung meningkat pada kualitas yang lebih tinggi, produsen dapat mengoptimalkan proses fermentasi untuk mencapai kadar alkohol yang lebih ideal dan mengontrol outlier dalam proses produksi.
Wine dengan kadar alkohol yang terlalu tinggi atau terlalu rendah perlu dievaluasi agar kualitas produk tetap konsisten.

## **HeatMap**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['alcohol', 'volatile acidity', 'quality']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Alkohol, Volatile Acidity, dan Kualitas Wine')
plt.show()
```

Alkohol adalah Faktor Penentu Terkuat: Terdapat hubungan positif sebesar 0.45 antara alcohol dan quality. Ini menunjukkan bahwa seiring meningkatnya kadar alkohol, kualitas wine cenderung dinilai lebih tinggi oleh para penguji.

Dampak Buruk Keasaman Volatil: Volatile acidity memiliki hubungan negatif yang cukup kuat, yaitu -0.38, terhadap quality. Artinya, semakin tinggi kadar asam volatil yang sering kali memberikan aroma cuka yang tajam, semakin rendah nilai kualitas wine tersebut.

Hubungan Lemah antara Alkohol dan Keasaman: hubungan antara alcohol dan volatile acidity sangat rendah dan negatif -0.19. Hal ini mengindikasikan bahwa kedua variabel ini bergerak secara independen; meningkatkan alkohol tidak secara otomatis menurunkan keasaman volatil, sehingga keduanya harus dikelola secara terpisah.

Fokus pada Pengendalian Keasaman:
Karena korelasi negatifnya signifikan -0.38, prioritas utama adalah menjaga agar volatile acidity tetap rendah melalui manajemen fermentasi yang bersih dan mencegah oksidasi.

Optimasi Target Kadar Alkohol:
